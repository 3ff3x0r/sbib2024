```r
# Suppress warnings
options(warn = -1)

# Extend the data
x <- c(41, 22, 20, 11, 50, 30, 60, 70, 25, 35)
y <- c(1, 4, 5, 1, 5, 2, 6, 3, 4, 7)
R <- c("A", "A", "A", "B", "B", "B", "C", "C", "C", "A")

# Define circle sizes (same length as x and y)
sizes <- c(2, 4, 6, 8, 10, 3, 7, 9, 5, 4)

# Map 'A', 'B', and 'C' to different colors
colors <- ifelse(R == "A", "red", ifelse(R == "B", "cyan", "green"))

# Plot with different circle colors based on the label in 'R'
symbols(x, y, circles = sizes, inches = 0.5, bg = colors, fg = "white", main = "Symbol Plot with Color by Label")

# Add labels for 'A', 'B', and 'C'
#text(x, y, labels = R, pos = 3)

# Add a legend
legend("topright", legend = c("A", "B", "C"), fill = c("red", "cyan", "green"), 
       title = "Labels", bty = "n")

```
